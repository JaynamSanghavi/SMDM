You will use the network of friends (G1) of an organization to answer questions in Part 1:<br>
Each node corresponds to a person, and an edge indicates their friendship.

```python
import networkx as nx
import operator
G = nx.Graph()
G = nx.read_gml('friends.gml')
#nx.draw(G, with_labels=True)
```

### Q1)Find the followings of node 100: <br>
 a) degree centrality<br>
 b) closeness centrality and<br>
 c) normalized betweeness centrality <br>

```python
degree_cen = nx.degree_centrality(G)
print('Degree Centrality : ' + str(degree_cen))
print(f"Degree centrality of node 100: {degree_cen[100]}")

closeness_cen = nx.closeness_centrality(G)
print('Closeness Centrality : ' + str(closeness_cen))
print(f"Closeness centrality of node 100: {closeness_cen[100]}")

betweenness_cen = nx.betweenness_centrality(G, normalized=True)
print('Normalized Betweeness Centrality : ' + str(betweenness_cen))
print(f"Normalized betweenness centrality of node 100: {betweenness_cen[100]}")


```

Q2) You are contracted by an emarketing company to send out a shopping voucher to a one
user in the network G1. Assume that user who receives the voucher will forward that it
his or her friends in the network G1. Select the best user that will be selected to send
the voucher, if the voucher has to reached as many nodes as possible and it can be
forwarded to multiple users at the same time, but the travel distance of the voucher is
limited to one step.

```python
# find the node with the highest degree centrality
max_dc_node = max(degree_cen, key=degree_cen.get)

print("The best user will be: ", max_dc_node)
```

```python
# find the node with the highest closeness centrality
max_cc_node = max(closeness_cen, key=closeness_cen.get)

print("The best candidate in the network to send voucher in lowest avg no of hops :", max_cc_node)
```
